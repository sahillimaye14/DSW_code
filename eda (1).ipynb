```python
#importing neccesary libraries
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
#reading the given csv
df = pd.read_csv("historic.csv")
# To display the top 5 rows
df.head(10)
```

```python
#converting the given data into binary format of 0 and 1
df['success_indicator'] = df['success_indicator'].apply(lambda x: 0 if x == 'flop' else 1)
```

```python
df.dtypes
```

```python
#Dropping unnecesary columns
df = df.drop(['item_no'], axis=1)
df.head(5)
```

```python
sns.violinplot(x = "main_promotion", y = "stars", data = df)
plt.show()
```

```python
import matplotlib.pyplot as plt
df['stars'].value_counts().sort_index().plot(kind='bar', figsize=(20, 4))
plt.xlabel('Stars')
plt.ylabel('Count')
plt.title('Frequency of Stars')
plt.show()
```

```python
df.groupby(['color', 'success_indicator'])['success_indicator'].count().unstack().plot(kind='bar')
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Success Indicator by Category')
plt.show()
```
